```python
import os

config = '/home/rubel/NOMAD-FAIRmat/GH/pynxtools-stm/tests/data/config_file_for_dat.json'
eln = '/home/rubel/NOMAD-FAIRmat/GH/pynxtools-stm/tests/data/eln_data.yaml'

def get_list_of_files(dir, ext):
    """
    Get the list of files with the given extension.
    """
    import os
    if not os.path.exists(dir):
        raise ValueError("Directory does not exist")
    print(os.path.abspath(dir))    
    files_out = []
    if os.path.isdir(dir):
        dir = os.path.abspath(dir)
    for root, dirs, files in os.walk(dir):
        for file in files:
            if file.endswith(ext):
                file = os.path.join(root, file)
                files_out.append(os.path.abspath(file))
    return files_out


def run_dataconverter_recursive(eln=None, config=None, 
                                file_ext=None, input_file=None, 
                                dir=None, max_file=3):
    input_files = []
    if input_file or dir:
        if input_file:
            input_files.append(input_file)
        else:
            input_files = get_list_of_files(dir, ext=file_ext)
    i = 0
    for file in input_files:
        if i>= max_file:
            return
        i = i + 1
        command = (f"dataconverter --input-file {file} --input-file {eln} --input-file {config} --reader {'sts'} --nxdl {'NXsts'} --output {file.replace(file_ext, 'nxs')}")
        os.system(command) 
```

```python
run_dataconverter_recursive(eln=eln, config=config, file_ext='dat', dir='./STS_for_template')
```

```python
import numpy as np
import h5py as h5
import copy
plot = {"f_value": None,
        "f_name": None,
        "f_unit": None,
        "a_values": [],
        "a_units": [],
        "a_names": [],
        "file_name": None,
        "plot_name": None}

plot_li = []

def get_plot_data(node, file_name="test_file", grp_name="grp_name", empty_plot=plot):
    plot = copy.deepcopy(empty_plot)
    plot["file_name"] = file_name
    plot_annotation = file_name
    # Add the functionality to see the default plot.
    # TODO collect the group name
    attrs = node.attrs
    axes = attrs.get("axes", [])
    if not len(axes):
        return {}
    field_key = ""
    for key, val in node.items():
        # field
        i = 1
        if key not in axes:
            if isinstance(val, h5.Dataset):
                field_key = key
                # TODO extend anotation for with group name
                plot_annotation = plot_annotation + "_fld( " + key + " )"
                plot["plot_name"] = plot_annotation
                plot["f_value"] = np.array(val)
                plot["f_name"] = key
                # field metadata
                for attr_k, attr_v in val.attrs.items():
                    if "units" == attr_k:
                        plot["f_unit"] = attr_v
        # attribute
        elif key in axes:
            plot["a_names"].append(key)
            plot["a_values"].append(np.array(val))
            i = i + 1
            for attr_k, attr_v in val.attrs.items():
                if attr_k == "units":
                    plot["a_units"].append(attr_v)
                
    return plot
    

def walk_through_NXdata(h5_node, node_name=""):
    """Walk through each node.
    """
    for key, val in h5_node.items():
        if isinstance(val, h5.Group):
            if val.attrs.get("NX_class", None) == "NXdata":
                if not node_name:
                    plot_li.append(get_plot_data(val, grp_name=key))
                elif key==node_name:
                    plot_li.append(get_plot_data(val, grp_name=key))
            else:
                walk_through_NXdata(val, node_name)
            
        
    

def read_and_store_plot_data(hdf5_file, node_name=""):
    """
    Extract and store plot data in h5 file.
    """

    if isinstance(hdf5_file, list) or isinstance(hdf5_file, tuple):
        for file in hdf5_file:
            read_and_store_plot_data(file, 
                                     node_name=node_name)
        return

    try:
        h5_file = h5.File(hdf5_file, "r+")    
        walk_through_NXdata(h5_file, node_name)
    finally:
        h5_file.close()
        return
    return

nxs_files = get_list_of_files(dir='./STS_for_template', ext='nxs')
read_and_store_plot_data(nxs_files, node_name='grad_current')
```

```python
import matplotlib.pyplot as plt
import numpy as np

tot_plt = len(plot_li)
nrow = int(np.ceil(tot_plt /3))

def plots(plt_dt_li, ncol=3, figsize = (8,6)):
    tot_plt = len(plot_li)
    if tot_plt < ncol:
        ncol = tot_plt
    nrow = int(np.ceil(tot_plt / ncol))
    fig, axes = plt.subplots(nrows=nrow,
                             ncols=ncol,
                             figsize=figsize
                            )
    for ind, plot_dct in enumerate(plt_dt_li):
        row_ind = int(ind/ncol)
        col_ind = int(ind % ncol)
        naxis = row_ind * ncol + ncol
        if ncol < tot_plt:
            naxis = (row_ind, col_ind)
        else:
            naxis = (ind)
            
        if tot_plt > 1:
            axes[naxis].plot(plot_dct["a_values"][0], 
                             plot_dct["f_value"])
            axes[naxis].set_title(plot_dct["plot_name"])
        else:
            axes.plot(plot_dct["a_values"][0], 
                             plot_dct["f_value"])
            axes.set_title(plot_dct["plot_name"])
            
    plt.tight_layout()
    plt.show()
    
                             
plots(plot_li, figsize=(40,40))
```
