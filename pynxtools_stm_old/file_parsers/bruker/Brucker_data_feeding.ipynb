```python
# Define Scan_control
def construct_multiple_scan_control_grps(template, conf_dict, parent_path: str, data_dict, layer_list=None):

    def construct_scan_control_grp(template, conf_dict, parent_path: str, data_dict, group_name='scan_control', layer_dict=None):
        """
        The function that constructs `scan_control` group. 
        """
        # layer_list is the list of dict. And each key in the dict is in byte literal and value is the list of byte literal
        
        # TODO: In app def scan_control should be upper case
        data_dct_k = conf_dict['/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]/scan_control/scan_region/scan_offset_x']
        scan_offset_x, unit_x = data_dict.get(data_dct_k, ' ').split(' ') 
        template[f"{parent_path}/{group_name}/scan_region/SCAN_OFFSET_N[scan_offset_x]"] = scan_offset_x 
        template[f"{parent_path}/{group_name}/scan_region/SCAN_OFFSET_N[scan_offset_x]/@units"] = BRUKER_UNIT_MAP.get(unit_x, unit_x)
        
        data_dct_k = conf_dict['/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]/scan_control/scan_region/scan_offset_y']
        scan_offset_y, unit_y = data_dict.get(data_dct_k, ' ').split(' ')
        template[f"{parent_path}/{group_name}/scan_region/SCAN_OFFSET_N[scan_offset_y]"] = scan_offset_y
        template[f"{parent_path}/{group_name}/scan_region/SCAN_OFFSET_N[scan_offset_y/@units]"] = BRUKER_UNIT_MAP.get(unit_x, unit_y)
    
        # Setting up the fast and slow axis:
        data_dct_k = conf_dict["/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]/scan_control/scan_region/scan_direction"]
        # TODO: what if layer_dict is none
        if data_dct_k.startswith('SCAN_LAYER/'):
            layer_dct_k = data_dct_k.split('SCAN_LAYER/')[-1]
            direction = layer_dict[layer_dct_k]
            if direction.lower() == 'down':
                template[f"{parent_path}/{group_name}/independent_scan_axes"] = ["-Y", "X"]

            elif direction.lower() == 'up':
                template[f"{parent_path}/{group_name}/independent_scan_axes"] = ["Y", "X"]
            elif direction.lower() == 'right':
                template[f"{parent_path}/{group_name}/independent_scan_axes"] = ["X", "Y"]
            elif direction.lower() == 'left':
                template[f"{parent_path}/{group_name}/independent_scan_axes"] = ["-X", "Y"]

        # Split the values into X and Y and put 
        scan_range = layer_dict['Scan Size']
        range_x, range_y, unit = scan_range.split(' ')
        range_x, range_y, unit = float(range_x), float(range_y), BRUKER_UNIT_MAP.get(unit, unit)
        
        template[f"{parent_path}/{group_name}/scan_region/SCAN_RANGE_N[scan_range_x]"] = range_x
        template[f"{parent_path}/{group_name}/scan_region/SCAN_RANGE_N[scan_range_x]/@units"] = unit 
        template[f"{parent_path}/{group_name}/scan_region/SCAN_RANGE_N[scan_range_y]"] = range_y
        template[f"{parent_path}/{group_name}/scan_region/SCAN_RANGE_N[scan_range_y]/@units"] = unit
    
        

        data_dict_key = conf_dict["/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]/scan_control/MESH_SCAN[mesh_scan]/scan_points_x"]
        if data_dict_key.startswith('SCAN_LAYER'):
            value_in_b = layer_dict[data_dict_key.split('SCAN_LAYER/')[-1]]
            # layer dict represents byte literal to list of bytes literal 
            value = int(value_in_b)
            print(' ###### scan name : ', scan_name)
            template[f"{parent_path}/{group_name}/MESH_SCAN[mesh_{scan_name}]/scan_pints_x"] = value
            template[f"{parent_path}/{group_name}/MESH_SCAN[mesh_{scan_name}]/step_size_x"] = range_x / value
            template[f"{parent_path}/{group_name}/MESH_SCAN[mesh_{scan_name}]/step_size_x/@units"] = unit
    
        data_dict_key = conf_dict["/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]/scan_control/MESH_SCAN[mesh_scan]/scan_points_y"]
        if data_dict_key.startswith('SCAN_LAYER'):
            value_in_b = layer_dict[data_dict_key.split('SCAN_LAYER/')[-1]]
            # layer dict represents byte literal to list of bytes literal 
            value = int(value_in_b)
                
            template[f"{parent_path}/{group_name}/MESH_SCAN[mesh_{scan_name}]/scan_pints_y"] = value
            template[f"{parent_path}/{group_name}/MESH_SCAN[mesh_{scan_name}]/step_size_x"] = range_y / value

    
    layer_list = [{key.decode(): value[0].decode() for key, value in x.items()} for x in layer_list]

    height_sensor_no = 0
    amplitude_error_no = 0
    phase_no = 0
    amplitude_no = 0
    tm_amplitude_no = 0

    for layer_info_dict in layer_list:
        # e.g. b'@2:Image Data': [b'S [ZSensor] "Height Sensor"'] 
        scan_name = layer_info_dict.get("@2:Image Data")
        # TODO: Test : Instead of the checking and creating 'Height Sensor, Amplitude Error and Phase
        # try to create module-level variables depending on the Scan type and late use that variable.
        if "Height Sensor" in scan_name:
            scan_name = f"height_sensor_{height_sensor_no}"
            height_sensor_no += 1;
        elif "Amplitude Error" in scan_name:
            scan_name = f"amplitude_error_{amplitude_error_no}"
            amplitude_error_no += 1
        elif "Phase" in scan_name:
            scan_name = f"phase_{phase_no}"
            phase_no += 1
        elif "Amplitude" in scan_name:
            scan_name = f"amplitude_{amplitude_no}"
            phase_no += 1
        elif "TM Deflection" in scan_name:
            scan_name = f"tm_deflection_{tm_amplitude_no}"
            phase_no += 1
        
        construct_scan_control_grp(template=template, conf_dict=conf_dict, 
                                   parent_path=parent_path, data_dict=data_dict, group_name=f"SCAN_CONTROL[scan_{scan_name}]",
                                   layer_dict=layer_info_dict)
                                      

def create_nexus_file():
    app_def = "NXafm"
    out_file = "/home/rubel/NOMAD-FAIRmat/GH/pynxtools-stm/pynxtools_stm/file_parsers/bruker/out.nxs"
    input_file = "/home/rubel/NOMAD-FAIRmat/GH/pynxtools-stm/tests/new_data/SPM_from_tamas/VGEP-15m-.0_00000.spm"

    from pynxtools_stm.file_parsers.bruker.brukerSPM import BrukerSPM
    import numpy as np

    from pynxtools.dataconverter.template import Template
    from pynxtools.dataconverter.helpers import generate_template_from_nxdl, get_nxdl_root_and_path

    from pynxtools.dataconverter.writer import Writer

    # Parse bruker file
    bkr = BrukerSPM(input_file)
    metadata = bkr.parsed_data
    
    # metadata_list = [(key, val) for key, val in metadata.items()]
    layer_list = bkr.layers # 8 layers
    scanner_list = bkr.scanners # 1 scanners
    channel_list = bkr.list_channels()
    
    
    template = Template()
    root, nxdl_path = get_nxdl_root_and_path(app_def)
    # generate_template_from_nxdl(root, template)
    

    construct_multiple_scan_control_grps(template, conf_dict=AFM_template, 
                               parent_path="/ENTRY[entry]/experiment_insturment/ENVIRONMENT[environment]",
                               data_dict=metadata,
                               layer_list=layer_list)
    Writer(template, nxdl_f_path=nxdl_path, output_path=out_file).write()
    print(template)
    
```
